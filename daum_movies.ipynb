# Daum 영화 이미지 받아오기

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://search.daum.net/search?w=tot&q=2019%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR"

res = requests.get(url)
res.raise_for_status()

soup = BeautifulSoup(res.text, "lxml")
```

```python
# image url 가져오기

images = soup.find_all("img", attrs={"class":"thumb_img"})

for image in images:
  if image["src"].startswith("//"):
    imageUrl = "https:" + image["src"]
    print(imageUrl)
  else:
    continue
```

```python
# 가져온 url을 가지고 이미지 다운받기

for index, image in enumerate(images):
  image_res = requests.get("https:" + image["src"])
  image_res.raise_for_status()

  with open("movie{}.jpg".format(index + 1), "wb") as f:
    f.write(image_res.content)
```
