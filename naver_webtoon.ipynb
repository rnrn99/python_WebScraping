```python
import requests
import re
from bs4 import BeautifulSoup
```

```python
url = "https://comic.naver.com/webtoon/weekday.nhn"

res = requests.get(url)
res.raise_for_status()

# res.text를 lxml parser를 이용해 BeautifulSoup 객체로 만든 후 soup에 저장
soup = BeautifulSoup(res.text, "lxml")
```

```python
print(soup.title)
```

# 웹툰 제목 가져오기

```python
# 전체 웹툰 제목 가져오기

cartoons = soup.find_all("a", attrs={"class":"title"})
for cartoon in cartoons:
  print(cartoon.get_text())
```

```python
# 목요웹툰 제목 가져오기

thuCartoons = soup.select('#content > div.list_area.daily_all > div:nth-of-type(4) > div > ul > li')

for thuCartoon in thuCartoons:
  thuCartoonName = thuCartoon.select_one('a.title')
  print(thuCartoonName.get_text())
```

# 모죠의일지 정보 가져오기

캡처.JPG

```python
url= "https://comic.naver.com/webtoon/list.nhn?titleId=728015&weekday=sat"
res = requests.get(url)
res.raise_for_status()
soup = BeautifulSoup(res.text,"lxml" )
```

```python
mojyo = soup.find_all("td", attrs={"class":"title"})
title = mojyo[0].a.get_text()
print(title)
```

```python
for item in mojyo:
  title = item.a.get_text()
  link = "https://comic.naver.com" + item.a["href"]
  print(title, link)
```

```python
totalNum = 0
rating = soup.find_all("div", attrs={"class":"rating_type"})

for item in rating:
  rate = item.find("strong").get_text()
  totalNum += float(rate)

print("Avg : ", totalNum / len(rating))
```
